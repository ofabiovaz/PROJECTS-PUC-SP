```python
#Bibliotecas usadas 
import pandas as pd
import numpy as np
```

```python
#csv
path = '/content/Ocorrências_da_lista_de_espécies.csv'
```

```python
#dataframe
df = pd.read_csv(path)
df.head()
```

```python
# transformando tudo que é nulo em vazio para organizar os uniques
df = df.replace(np.nan, ' ')
```

```python
#uniques dos state em ordem alfabética para melhor observação
sorted(df['state'].unique())
```

```python
#sp Padrão
sp = 'São Paulo'
# Lista com sp diferentes
sp_state_list = ['SÃ£o Paulo','SP','São paulo','S.Paulo','S�o Paulo','S  o Paulo']
# replace dos sp diferentes para sp padrão
df = df.replace(sp_state_list,sp)
```

```python
#uniques dos state com a tranformação
sorted((df['state'].unique()))
```

```python
#df com state de São Paulo
df_state_sp = df[df['state']=='São Paulo']
#uniques dos city em ordem alfabética para melhor observação
sorted(df_state_sp['city'].unique())
```

```python
#lista com city diferentes 
sp_city_list = ['Sao Paulo','SÃO PAULO',' São Paulo','São Paul']

#transformação dos city diferentes em padrão 
df = df.replace(sp_city_list,sp)
```

```python
#uniques city após a transformação 
sorted(df_state_sp['city'].unique())
```

```python
#df estado e cidade de São Paulo
df_state_city_sp = df[df['state']=='São Paulo'][df['city']=='São Paulo']
df_state_city_sp
```
